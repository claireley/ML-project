```python
import pandas as pd
import numpy as np
import yaml
```

```python
with open("../config.yaml", "r") as file:
    config = yaml.safe_load(file)
```

```python
def clean_col_names(df):
    df2 = df.copy()
    df2.columns = df2.columns.map(lambda x: x.lower().replace(' ',''))
    return df2

def remove_empty_fields(df):
    df2 = df.copy()
    empty = df2[df2.iloc[:,-1]==9].index
    df2.iloc[empty,-2]= np.nan
    df2 = df2.interpolate()
    return df2

def remove_leap_days(df):
    df2 = df.copy()
    df2 = df2.sort_values('date', ascending=True)
    df2.date  = pd.to_datetime(df2.date, format='%Y%m%d')
    df_filtered = df2[df2['date'].dt.strftime('%m-%d') == '02-29']
    df2 = df2.drop(df_filtered.index.tolist())
    df2 = df2.reset_index(drop=True)
    return df2

def drop_columns(df):
    df2 = df.copy()
    if df2.columns[0]=='souid':
        df2 = df2.drop(columns = [df.columns[0]])
    if df2.columns[-1][0]=='q':
        df2 = df2.drop(columns = [df.columns[-1]])
    return df2

def filter_years(df):
    df2 = df.copy()
    df2['year']=df2.date.dt.year
    df2 = df2[df2.year >= 1991]
    return df2

def clean_data(df):
    df2= df.copy()
    df2 = clean_col_names(df2)
    df2 = remove_leap_days(df2)
    df2 = remove_empty_fields(df2)
    if sum(df2.iloc[:,-1]==1) == 0:
        df2 = drop_columns(df2)    
    else:
        print('has missing values')
    #df2 = filter_years(df2)
    return df2
```

```python
df_cc = pd.read_csv(config['data']['raw']['file1'], quotechar='"', sep = ";")
df_cc = clean_data(df_cc)
df_cc.to_csv("../data/clean/cc.csv", index=False, encoding= "utf-8", sep = ";")
```
